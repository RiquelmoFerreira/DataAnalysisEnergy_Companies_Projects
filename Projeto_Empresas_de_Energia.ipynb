## Análise do valor das ações das companias brasileiras do Setor de Energia

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

### Dataset

```python
Data_Set = pd.read_excel('Dados_Empresas_Energia.xlsx')

Data_Set.head()
```

### Renomeando

```python
Data_Set.rename(columns={'Data':'Date'}, inplace=True)
```

### Definindo o Index

```python
Data_Set.set_index('Date', inplace=True)

Data_Set.head()
```

### Gráfico de Linha

```python
plt.figure(figsize=(16, 7))

plt.title('Energy Companies Shares Analysis', loc='left', fontsize=18)

#Petrobras
plt.plot(Data_Set.index, Data_Set['Petrobras'], color='#008c4a', linewidth=4, alpha=0.7)
plt.text(Data_Set.index[-1], Data_Set['Petrobras'].tail(1), 'Petrobras', color='#008c4a',
        size='large', horizontalalignment='left'
        )

#Others Companies
colors = ['#12FF11', '#E0302C', '#F7921E', '#E0109F', '#113EF8']
count = 0
for column in Data_Set.columns[1:]:
    plt.plot(Data_Set.index, Data_Set[column], color=colors[count], linewidth=2, alpha=0.5)
    plt.text(Data_Set.index[-1], Data_Set[column].tail(1), column, color=colors[count],
    size='large', horizontalalignment='left'
    )  
    count += 1
plt.xlabel('Period')
plt.ylabel('Closing Price (R$)');
```
